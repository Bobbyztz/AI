```python
print('--------------------- one dim vector ---------------------')
```

```python
x1 = tc.tensor([1])
print('x1 = ', x1)
print('x1.shape = ', x1.shape)
print('x1.numel() = ', x1.numel())
print('x1.ndimension() = ', x1.ndimension())
```

```python
print('--------------------- vector ---------------------')
```

```python
y = tc.tensor([1, 3, 5, 7])
print('y = ', y)
print('y.shape = ', y.shape)
print('y.numel() = ', y.numel())
print('y.ndimension() = ', y.ndimension())
```

### tc.tensor([1]) is different with tc.tensor(1) in shape,dim and usage

```python
print('--------------------- check cuda ---------------------')
```

```python
print('y.device = ', y.device)
if tc.cuda.is_available():
    print('CUDA available. Put x to CUDA...')
    y = y.to(device='cuda')
    print('y.device = ', y.device)
```

```python
print('--------------------- vector slice ---------------------')
```

```python
print('y[2] = ', y[2])
print('y[-1] = ', y[-1])
print('y[1:3] = ', y[1:3])
print('y[1:] = ', y[1:])
print('y[:3] = ', y[:3])
print('y[:] = ', y[:])
```

```python
print('--------------------- matrix ---------------------')
```

```python
I = tc.tensor([[1, 0], [0, 1]]) # transfer list into matrix/tensor
print('I = \n', I)
print('I.ndimension() = ', I.ndimension())
print('I.shape = ', I.shape)
```

```python
I1 = tc.eye(2) # identity matrix
print('I1 = \n', I1)
```

```python
print('--------------------- memory mechanism ---------------------')
```

```python
pauli_z = I1.clone() # clone()
pauli_z[1, 1] = -1
print('I1 = \n', I1)
print('pauli_z = \n', pauli_z)
```

```python
K = I1 * 1 # "="+( )
K[1, 1] = -1
print('I1 = \n', I1)
print('K = \n', K)
```

```python
J = I # "="
print('I = ', I)
J[1, 1] = -1
print('I = \n', I)
print('J = \n', J)
```

```python
print('--------------------- matrix slice ---------------------')
```

```python
z = tc.randn((4, 4))
print('z = \n', z)
print('z[1, 1:3] = ', z[1, 1:3])
print('z[1, 1:] = ', z[1, 1:])
print('z[1, :3] = ', z[1, :3])
print('z[1, :] = ', z[1, :])
print('z[1:3, :2] = \n', z[1:3, :2])
```

```python
print('--------------------- Higher order tensor ---------------------')
```

```python
print('# delta tensor')
delta = tc.tensor([[[1, 0], [0, 0]], [[0, 0], [0, 1]]])
print('delta = \n', delta)
print('delta.ndimension() = ', delta.ndimension())
print('delta.shape = ', delta.shape)
```

```python
print('# randomn tensor')
w = tc.randn((2, 2, 2), device='cpu', dtype=tc.float64)
print('w = \n', w)
```
