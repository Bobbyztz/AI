```python
import torch as tc

x = tc.tensor([1, 3, 5, 7])
x1 = tc.reshape(x, [2, 2])
x1_ = x.reshape(2, 2)
print(x1)
print(x1_)

# Test index of element "1"
print('Test index of element "1"', x[0], x1[0, 0])
# Test index of element "5"
print('Test index of element "5"', x[2], x1[1, 0])
```

```python
print('--------------------- Matrix transpose ---------------------')
```

```python
x2 = x1.permute(1, 0)
# Test index of element "5" before permute
print('Test index of element "5" before permute', x1[1, 0])
# Test index of element "5" after permute
print('Test index of element "5" after permute', x2[0, 1])
x2_ = x1.t()
print('x2 = ', x2)
print('x2_ = ', x2_)
```

```python
print('--------------------- vectorization by flatten or reshape ---------------------')
```

```python

T = tc.randn(2, 3, 4)
# Shape of tensor T
print('Shape of tensor T', T.shape)
V = T.flatten()
# Shape after flatten
print('Shape after flatten', V.shape)

# Difference between results of flatten and reshape
# norm() - Computes the Euclidean norm (L2 norm) of a tensor
# item() - Converts a one-element tensor to a Python scalar
print('Difference between flatten and reshape results', (V - T.reshape(-1)).norm().item())
# when only doing (V - T.reshape(-1)).norm(), the returned tensor(.0) means
# A PyTorch tensor with a single scalar value 0.0.
# It means the L2 norm of the difference between V and T.reshape(-1) is zero, indicating they are identical.

# Shape of matrixized T[0,1]. '-1' here means calcualte the dim automatically
print('Shape of matrixized T[0,1]', T.reshape(-1, 4).shape)
```
