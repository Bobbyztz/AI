```python
import torch as tc
```

```python

u = tc.randn(4)
v = tc.randn(4)

z1 = u.dot(v)
z2 = u.inner(v)
z3 = tc.einsum('n,n->', u, v)
print('Results of vector inner product calculated by three methods:', z1, z2, z3)




```

```python
z4 = 0
for n in range(u.numel()):
    z4 += u[n] * v[n]
print('Result of vector inner product calculated by loop:', z4)
```

```python
z5 = (u * v).sum()
print('Result of vector inner product calculated by Hadamard product and sum:', z5)
```
