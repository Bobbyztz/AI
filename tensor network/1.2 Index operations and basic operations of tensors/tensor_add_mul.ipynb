```python
import torch as tc
```

```python
u = tc.tensor([1, 2, 3, 4])
v = tc.tensor([2, 4, 6, 8])
c = 2
d = tc.tensor([2])
```

```python
print('u * c = ', u * c)
print('u * d = ', u * d)
print('u + c = ', u + c)
print('u + v = ', u + v)
print('u * v = ', u * v)
```

```python
print('\nOuter product of vectors (outer) = \n', u.outer(v))
print('Outer product of vectors (einsum) = \n', tc.einsum('a,b->ab', u, v))

print('\nKronecker product of vectors (kron) = ', tc.kron(u, v))
print('Kronecker product of vectors (einsum) = ', tc.einsum('m,n->mn', u, v).reshape(-1, ))
```

```python
m1 = tc.randn(2, 2)
m2 = tc.randn(2, 2)
print('\nDifference between Kronecker product (kron) and einsum = \n',
      tc.kron(m1, m2) - tc.einsum('ab,cd->acbd', m1, m2).reshape(4, 4))
```
